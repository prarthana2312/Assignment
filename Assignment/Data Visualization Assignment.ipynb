## 1.	Did the piece rate motivate workers?

```python
import warnings 
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
sns.factorplot(x="Worker", y="Output",data=df,size=4, aspect=4)
```

```python
fig = plt.figure(figsize=(8,4), dpi=100)
sns.lineplot(x="Worker", y="Output",data=df)
```

#### Yes, PieceRate motivates workers to work ever harder for their earnings. As in this 2 plots there are increase in production of electrodes by workers.
#### In the first plot we can see some workers electrode production is good. eg: Worker(6th) has produced electrodes between 500-700, Worker(45th) has produced electrodes  between 900-1000.
#### The second plot shows the production of electrodes by workers.

## 2. Did it attract workers of higher ability?

```python
fig = plt.figure(figsize=(10,5))
sns.boxplot(x="Worker", y="Week",data=df)
```

```python
fig = plt.figure(figsize=(8,4), dpi=100)
sns.lineplot(x="Worker", y="Week",data=df)
```

```python
fig = plt.figure(figsize=(8,4), dpi=100)
sns.scatterplot(x="Worker", y="Week",hue="Output",data=df)
```

#### Yes,PieceRate has attract some workers with higher ability. We can see in the first plot Worker 2nd and Worker 10th has worked from Week 1 till Week 50. and some worker have started in the middle of the week till the end.
####  Second Plot shows the plot of the workers.
#### Third Plot shows the Worker working in the week and their output.

## 3. Did the introduction of the piece rate make workers feel special because they were part of an experiment and wanted to help the managers to succeed in the experiment? That is, is there any evidence that workers enthusiasm was temporary?

```python
fig = plt.figure(figsize=(12,6), dpi=100)
sns.scatterplot(x="Worker",y="Week",data=df)
```

#### Yes, there is an evidence that workers enthusiasm was temporary. As we can see in this only plot 2nd and 10th worker is working till the end of week i.e(50) and most of the workers have worked for few weeks. 

## 4. Did workers learn on the job (that is, get better with experience) and, if so, how should this affect conclusions about the effects of piece rates?



```python
fig = plt.figure(figsize=(8,4), dpi=100)
sns.lineplot(x="Week", y="Output",data=df)
```

```python
sns.relplot(x="Week", y="Output",hue="Worker",legend="full",height=5,aspect=2,data=df)
```

####  In the first plot we can see that the production of the electrodes has increased during the particular time i.e after 26th week ,so workers were getting experience and were getting better.
#### Second Plot shows the output in a week by every worker.
